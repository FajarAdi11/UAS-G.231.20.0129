```python
# import libraries

from ultralytics import YOLO
import yaml
import cv2
from google.colab.patches import cv2_imshow
```

```python
import matplotlib.pyplot as plt
import time
%matplotlib inline
def convertToRGB(img):
 return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
test1 = cv2.imread('/content/drive/MyDrive/UAS AI/BARENG.jpg')
gray_img = cv2.cvtColor(test1, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
```

```python
# instansiasi kelas YOLO

model = YOLO("yolov8n.pt")
```

```python
model.predict("/content/drive/MyDrive/UAS AI/BARENG.jpg", save = True, save_txt = False)
```

```python
import cv2
haar_face_cascade = cv2.CascadeClassifier('/content/drive/MyDrive/UAS AI/haarcascade_frontalface_alt.xml')
faces = haar_face_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5)
print('Faces found: ', len(faces))
for (x, y, w, h) in faces:
  cv2.rectangle(test1, (x, y), (x+w, y+h), (0, 255, 0), 2)
plt.imshow(convertToRGB(test1))
```
